Exercitiul 1


```python
t = np.linspace(0, 0.03, int(0.03/0.0005) + 1)
```

```python
x_t = np.cos(520 * np.pi * t + np.pi/3)
y_t = np.cos(280 * np.pi * t - np.pi/3)
z_t = np.cos(120 * np.pi * t + np.pi/3)

```

```python
fig, axs = plt.subplots(3)
fig.suptitle('Exercitiul 1')

# Subplot pentru x(t)
axs[0].plot(t, x_t)
axs[0].set_ylabel('x(t)', fontsize=12)
axs[0].set_title('x(t) = cos(520pit + pi/3)')

# Subplot pentru y(t)
axs[1].plot(t, y_t)
axs[1].set_ylabel('y(t)', fontsize=12)
axs[1].set_title('y(t) = cos(280pit - pi/3)')

# Subplot pentru z(t)
axs[2].plot(t, z_t)
axs[2].set_xlabel('Timp (s)', fontsize=12)
axs[2].set_ylabel('z(t)', fontsize=12)
axs[2].set_title('z(t) = cos(120pit + pi/3)')

for ax in axs.flat:
    ax.set_xlim([0, 0.03])
    ax.set_ylim([-1.5, 1.5])

plt.tight_layout()
plt.show()
```

```python
frec=200
perioada=1/frec
print("Perioada semnalului: ", perioada)



n = np.arange(0, 0.03, perioada)

x_n = np.cos(520 * np.pi * n + np.pi/3)
y_n = np.cos(280 * np.pi * n - np.pi/3)
z_n = np.cos(120 * np.pi * n + np.pi/3)


fig, axs = plt.subplots(3)
fig.suptitle('Semnale esantionate la 200hz')

axs[0].stem(n, x_n)
axs[0].set_title('x[n]')

axs[1].stem(n, y_n)
axs[1].set_title('y[n]')

axs[2].stem(n, z_n)
axs[2].set_title('z[n]')
```

Un semnal sinusoidal de frecven 400 Hz, care s con-
in 1600 de eantioane.

```python
frec=400
n=1600

t=np.linspace(0, 4, 1600)

x= np.sin(2 * np.pi * frec * t)

fig, ax = plt.subplots()
ax.plot(t, x)
plt.show()
```

Un semnal sinusoidal de frecven 800 Hz, care s dureze
3 secunde.

```python
frec=800

t=np.linspace(0, 3, 1600)

y= np.sin(2 * np.pi * frec * t)

fig, ax = plt.subplots()
ax.plot(t, y)
plt.show()
```

Un semnal de tip sawtooth de frecven 240 Hz (putei
folosi funciile np.floor sau np.mod). 

```python
frec=240
n= 100

t = np.linspace(0, 0.01, n) 

semnal = 2 * (t * frec - np.floor(0.5 + t * frec))

plt.plot(t, semnal)
plt.grid(True)
plt.show()
```

Un semnal de tip square de frecven 300 Hz (putei folosi
funcia np.sign). 

```python
frec= 300
n= 10000

t = np.linspace(0, 0.01, n)

s = np.sign(np.sin(2 * np.pi * frec * t))

plt.plot(t, s)
plt.show()
```
